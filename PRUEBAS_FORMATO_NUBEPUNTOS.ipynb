```python
with open("captura_sin_ultimas_coordenadas.xyz", "r") as f:
    # Lee las líneas del archivo
    lines = f.readlines()

    # Crea una lista para almacenar las coordenadas
    coordenadas = []

    # Itera sobre las líneas y divide las coordenadas
    for line in lines:
      # Divide la línea en cinco partes
      x, y, z = line.split()

      # Almacena las primeras tres coordenadas
      coordenadas.append([float(x), float(y), float(z)])

    # Imprime las primeras 5 coordenadas
    print(coordenadas[:5])
    print(len(coordenadas))
    print(type(coordenadas))
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

for coord in enumerate(coordenadas[:50000]):
    ax.scatter3D(coord[0], coord[1], coord[2])

# Set labels and title
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.set_title('Point Cloud Visualization')

# Show the plot
plt.show()
```
